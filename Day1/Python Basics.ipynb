# List Comprehension

Create a list based on a logic,rule, any linear syntax

```python
# Create  a list of squares of numbers from 1 - 6
lst = []
for i in range(6):
    lst.append(i**2)

print(lst)
```

```python
squares = [i**2 for i in range(6)]
print(squares)
```

```python
# prompt: Create a list of store and calculate avaerage marks of students and also print their max, min, and average

# Sample student data (replace with your actual data)
student_data = {
    "Bharath": [85, 92, 78, 95, 88],
    "Kumar": [70, 80, 75, 85, 90],
    "Manoj": [90, 95, 88, 92, 98],
    "Satish": [78, 85, 80, 90, 75]
}

# Calculate average marks, max, min for each student
for student, marks in student_data.items():
  average_marks = sum(marks) / len(marks)
  max_marks = max(marks)
  min_marks = min(marks)

  print(f"Student: {student}")
  print(f"  Average Marks: {average_marks}")
  print(f"  Maximum Marks: {max_marks}")
  print(f"  Minimum Marks: {min_marks}")
```
